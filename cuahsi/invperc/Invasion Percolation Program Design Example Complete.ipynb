All our work in one file 

```python
# The complete solution
import numpy as np
from random import random
from ipythonblocks import BlockGrid

# Function to create resistance grid
def create_resistance_grid(nx,ny):
    resistance=np.zeros([ny,nx])
    for y in range(ny):
        for x in range(nx):
            resistance[y,x]=random()
    return resistance

# Function to initialize filled grid
def initialize_filled_grid(nx,ny,startx,starty):
    filled=np.zeros([ny,nx])
    filled[starty,startx]=1
    return filled

# Function to percolate one grid cell
def percolate(resistance,filled):
    found=[-1,-1]  # to keep track of the cell found with minimum resistance
    foundValue=1.1  # The minimum found value, relying on random numbers being between 0 and 1
    nsides=4  # The number of side cells
    xsteps=[1,0,-1,0]
    ysteps=[0,1,0,-1]
    ysize,xsize=filled.shape  # Would be good to check that resistance and filled have consistent shape
    for y in range(ysize):
        for x in range(xsize):
            if(filled[y,x]==1):
                for side in range(nsides):
                    # get the neighbor cells
                    xn=x+xsteps[side]
                    yn=y+ysteps[side]
                    # check if in the grid
                    if(xn < xsize and xn>=0 
                        and yn < ysize and yn >=0
                        and filled[yn,xn]==0  # Has not been percolated
                        and resistance[yn,xn]<foundValue):  # Here we have a candidate cell to percolate into
                            found=[yn,xn]
                            foundValue=resistance[yn,xn]
    if(found[0]>=0 and found[1]>=0):  #  Here we did actually find a cell to percolate to
        filled[found[0],found[1]]=1
    return filled


# Function to return block plot
def blockplot(filled,startx,starty,bs):
    ysize,xsize=filled.shape
    bgrid=BlockGrid(xsize,ysize,fill=(123,234,123),block_size=bs)    # note that number of columns needs to be given before number of rows in creating block grid object
    red=(255,0,0)
    for x in range(xsize):
        for y in range(ysize):
            if(filled[y,x]==1): 
                bgrid[y,x].set_colors(*red)
    bgrid[starty,startx].set_colors(0,0,0)
    return bgrid


#  Now we can set our inputs and do the work
nx=100
ny=80
startx=50  # note indexing from 0
starty=40
nsteps=600
resistance=create_resistance_grid(nx,ny)
filled=initialize_filled_grid(nx,ny,startx,starty)
for i in range(nsteps):
    filled=percolate(resistance,filled)
pgrid=blockplot(filled,startx,starty,5)
pgrid
```
